```python
# Pairplots - Can be used to see the relationship and distribution of more than 2 variables at a time

newdf=df[['Age','Glucose','Pregnancies']]
sns.pairplot(data=newdf)
```

```python
# 3D PLOT

fig = plt.figure(figsize = (15, 8))
ax = fig.add_subplot(projection='3d')
ax.scatter(xs=df["Age"], ys=df["Glucose"],zs=df["Pregnancies"]) #Plot scatter
ax.set_xlabel("Age")
ax.set_ylabel("Glucose")

ax.set_zlabel("Pregnancies");

plt.show()
```
