```python
# Step 1 : Importing the library

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Lets plot the basic simple graph

plt.plot([1,2,3],[4,5,2])

# To show the plot

plt.show()
```

```python
# Lets add title and labels to our graph

Marks=[60,75,80,90]
Grade = ['C','B','A','A+']

plt.plot(Marks,Grade)
plt.title("Students Info")
plt.xlabel("Student Marks")
plt.ylabel("Students Grade")
plt.show()
```

```python
# To include color, change the width and add grids to the graph, with the use of ggplot we can achieve that

from matplotlib import style
style.use('ggplot')

A=[5,8,9]
B=[12,16,6]

A1=[14,15,16]
B1 =[30,35,15]

plt.plot(A,B,'g',label='First plot',linewidth=5)
plt.plot(A1,B1,'c',label='second plot',linewidth=5)

plt.title("My Data")
plt.xlabel("Age")
plt.ylabel("Weight")

plt.legend()
plt.grid(True,color='k')
plt.show()
```

```python
# Bar plot

X=[5,7,10]
Y=[10,12,14]

A=[9,19,21]
B=[25,30,45]

plt.bar(X,Y,label="First Set")
plt.bar(A,B,label="Second Set",color='g')

plt.title("Random Data")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.legend()
plt.show()
```

```python
# Bar plot especially used for categorical or qualitative data

X=['Sandeep','Mallesh','Yamin']
Y=[30,45,70]

plt.bar(X,Y)
plt.show()
```

```python
# Scatter plot - It is used to understand the correlation between variables(Positive,Negative,NoCorrelation)
# Positive correlation

AmountSpent=[1000,2000,2500,3000,4000,10000]
Profit=[7000,8000,8500,11000,15000,25000]

plt.scatter(AmountSpent,Profit)
plt.title("Correlation")
plt.xlabel("Amount In Rs")
plt.ylabel("Profit Recieved")

plt.show()
```

```python
# Scatter plot - Negative correlation

AmountSpent=[1000,2000,2500,3000,4000,10000]
Profit=[7000,4000,3500,2000,1500,500]

plt.scatter(AmountSpent,Profit)
plt.title("Correlation")
plt.xlabel("Amount In Rs")
plt.ylabel("Profit Recieved")

plt.show()
```

```python
# Scatter plot - No correlarion

AmountSpent=[1000,2000,2500,3000,4000,10000]
Profit=[100,100,100,100,100,100]

plt.scatter(AmountSpent,Profit)
plt.title("Correlation")
plt.xlabel("Amount In Rs")
plt.ylabel("Profit Recieved")

plt.show()
```
